```python
cmap = 'PuBuGn'

def expected_Aija(U,V,W):
    if W.ndim==1:
        M=np.einsum('ik,jk->ijk',U,V)
        M=np.einsum('ijk,k->ij',M,W)
    else:    
        M=np.einsum('ik,jq->ijkq',U,V)
        M=np.einsum('ijkq,kq->ij',M,W)
    return M
```

```python
label = '500_3_20.0_0.0_False_0'
label = 'syn_'+label 
```

```python
dir0 = '../data/input/'
dir1 = dir0 + 'synthetic/edge_anomalies/' 
```

```python
# # original parameters
theta = np.load(dir1+'theta_'+label.split('syn_')[-1]+'.npz',allow_pickle=True)
# # define some support files
np.savetxt("../data/input/u.dat", theta['u'], delimiter=" ")
np.savetxt("../data/input/v.dat", theta['v'], delimiter=" ") 

theta['mu'],theta['pi'],theta['u'].shape
```

```python
network = dir1 +label+'.dat'


A, B, B_T, data_T_vals = tl.import_data(network,header=0)
nodes = A[0].nodes()
pos = nx.spring_layout(A[0]) 
N = len(nodes)
L = B.shape[0]
K = theta['u'].shape[-1]
print(N)
```

### Initializing near ground truth

```python
fold = 1
prng = np.random.RandomState(seed=10)
rseed = prng.randint(1000)
indices = cvfun.shuffle_indices_all_matrix(N, L, rseed=rseed)
mask = cvfun.extract_mask_kfold(indices, N, fold=fold, NFold=5)
np.count_nonzero(theta['z'][mask[0]])
```

```python
B_train = B.copy()
B_train[mask > 0] = 0
```

```python
%%time
flag_anomaly = True
mask_input = np.logical_not(mask)
# mask_input = None
model = adet.AnomalyDetection(plot_loglik=True,verbose = 1,initialization=0,K=K,N=N,N_real=1,
                              undirected=False,fix_communities=False,flag_anomaly=flag_anomaly,
                              in_parameters=dir1+'theta_'+label.split('syn_')[-1],
                              fix_pibr=False, fix_mupr=False,bg=10.,out_inference=False,
                            end_file = label,constrained=False,assortative=True)

u,v,w,pi,mu,maxL = model.fit(data=B_train,nodes=nodes,mask = mask_input)
 
print('\nInferred mu: {}, true mu: {}.\n'.format(mu,theta['mu']))
print('\nInferred pi: {}, true pi: {}.\n'.format(pi,theta['pi']))
```

### AUC
- $(A,\hat{A})$
- $(Q,Z)$

```python
M0 = cvfun._lambda0_full(u, v, w)
if flag_anomaly == False:
    Q = np.zeros_like(M0)
else:
    Q = cvfun.calculate_Q_dense(B,M0,pi,mu)
M = cvfun.calculate_expectation(u,v,w,Q[0],pi)

Q = cvfun.calculate_Q_dense(B,M0,pi,mu)
```

```python
maskQ = np.logical_and(mask[0]>0,B[0]>0)
cvfun.calculate_AUC(M[0], B[0], mask=mask[0]),cvfun.calculate_AUC(Q[0], theta['z'], mask=maskQ)
```

```python
f1 = cvfun.calculate_f1_score(Q,theta['z'], mask = maskQ)
f1
```
